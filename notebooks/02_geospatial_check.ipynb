```python
from pathlib import Path
import rasterio

print(rasterio.__version__)

PROJECT_ROOT = Path.cwd().parent

MS_DIR = (
    PROJECT_ROOT
    / "data"
    / "eurosat"
    / "EuroSATallBands"
)
```

```python
sample_file = next(
    MS_DIR.rglob("*.tif")
)

print(sample_file)
```

```python
with rasterio.open(sample_file) as src:

    print("CRS:")
    print(src.crs)

    print("\nBounds:")
    print(src.bounds)

    print("\nWidth:", src.width)
    print("Height:", src.height)

    print("\nBand Count:")
    print(src.count)
```

```python
from rasterio.warp import transform_bounds

with rasterio.open(sample_file) as src:

    bounds = transform_bounds(
        src.crs,
        "EPSG:4326",
        *src.bounds
    )

    print(bounds)
```
